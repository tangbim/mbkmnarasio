## **1.1. Function**
Fungsi adalah blok kode yang hanya berjalan saat dipanggil. \\
Anda dapat mengirimkan data yang dikenal sebagai parameter, ke dalam suatu fungsi.

```python
def tambah(a, b):
    c = a + b
    return c
```

### **1.1.3 Fungsi dengan Parameter**

```python
def my_children(first_name): # function dengan 1 parameter
    print(first_name + " " + "Wijaya")
```

```python
my_children("Theo")
my_children("Tilda")
my_children("Tina")
```

```python
def my_function(child3, child2, child1):
  print("The youngest child is " + child3)

my_function(child1 = "Emil", child2 = "Tobias", child3 = "Linus")
```

> **Argumen Arbritary untuk menyelamatkan ketika kita tidak tahu caranya banyak argumen yang akan diteruskan ke fungsi Anda,  tambahkan * sebelum nama parameter dalam definisi fungsi.**

```python
def my_children(*first_name): # function dengan multiple parameter
    for nama in first_name:
        print(nama + " " + "Wijaya")
```

```python
my_children("Theo", "Tilda", "Tina")
```

```python
my_children("Andi", "Rudi")
```

```python
# Contoh fungsi barisan fibonacci
def fibonacci(N): 
    L=[]
    a,b=0,1 
    while len(L) < N:
        a,b=b,a+b
        L.append(a) 
    return L
```

```python
fibonacci(10)
```

### **1.1.4 Fungsi dengan Return Statement**

```python
# Membuat fungsi luas segitiga
def luas_segitiga(alas, tinggi): # alas dan tinggi merupakan parameter yang masuk
    luas = (alas * tinggi) / 2
    return luas
```

```python
# Pemanggilan fungsi
# 4 dan 6 merupakan parameter yang diinputkan kedalam fungsi luas segitiga 
luas_segitiga(4, 6)
```

### **1.1.5 Fungsi di dalam Fungsi**

```python
def fungsi1():
    s = 'Selamat Datang'

    def fungsi2():
        print(s)

    fungsi2()

# Pemaingglan Fungsi
fungsi1()
```

```python
def penjumlahan(angka):
  tambah = 1 + int(angka)
  
  def kali():
    dikali = int(tambah) * 2
    return dikali
    
  return kali()

# Pemanggilan fungsi
penjumlahan(3)
```

```python
def perkalian(a,b):
    return (a*b)

def luas_segitiga(n,m):
    luas = perkalian(n,m)/2
    return luas

# Pemanggilan
luas_segitiga(4,6)
```

### **1.1.6 Docstring dalam Fungsi**

```python
def ganjilgenap(x):
    """Fungsi ini untuk melakukan cek angka ganjil atau genap"""

    if (x % 2 == 0):
        print("Genap")
    else:
        print("Ganjil")
```

### **1.2. Lambda**

```python
# Fungsi penjumlahan
add = lambda x,y: x+y
add(3,4)
```

```python
# Fungsi Luas persegi panjang
luas_persegi_panjang = lambda panjang, tinggi: panjang*tinggi 
luas_persegi_panjang(4, 2)
```

```python
# Contoh lambda function pakai if else
max = lambda a, b : a if(a > b) else b
 
print(max(1, 2))
print(max(10, 2))
```

```python
# check jika angka yang masuk ganjil atau genap
result = lambda x : f"{x} adalah genap" if x %2==0 else f"{x} adalah ganjil"
  
# print for numbers
print(result(12))
print(result(11))
```

```python
# check jika dua angka lebih besar atau lebih kecil
result = lambda x,y : f"{x} is smaller than {y}" \
if x < y else (f"{x} is greater than {y}" if x > y \
               else f"{x} is equal to {y}")
  
  
# print for numbers
print(result(12, 11))
print(result(12, 12))
print(result(12, 13))
```

```python
greeting = lambda name: print(f"Hello, {name}") 
```

Karena fungsi lambda tidak punya nama, jadi kita butuh variabel untuk menyimpannya.

Nanti saat kita mau panggil, kita tinggal tuliskan saja nama variabelnya seperti ini:

```python
greeting("Petani Kode")
greeting("Dian")
greeting("Ayu")
```
