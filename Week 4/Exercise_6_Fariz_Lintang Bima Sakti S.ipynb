## Case 5

## Case 7
Kamu diminta untuk memberikan 5 data teratas yang berisi informasi mengenai *report_code, month, day, thunder, dan tornado*. Data yang kamu berikan akan membantu pusat informasi dalam station untuk memberikan peringatan pada para penumpang. Bagaimana cara mendapatkan informasi tersebut?

```python
clasdata = pd.read_sql_query("SELECT REPORT_CODE, MONTH, DAY, THUNDER, TORNADO FROM STATION_DATA ORDER BY YEAR DESC LIMIT 5",conn)
clasdata
```

## Case 8
Kapan terdapat laporan mengenai adanya thunder atau tornado ? urutkan waktu kejadian dari yang terbaru

```python
tornado = pd.read_sql_query("SELECT STATION_NUMBER,REPORT_CODE,YEAR,MONTH,DAY,THUNDER,TORNADO FROM STATION_DATA WHERE THUNDER>0 OR TORNADO>0 ORDER BY YEAR DESC",conn)
tornado
```

## Case 9
Berapa kali terjadinya tornado dan rata-rata wind_speed & temperatur setiap tahunnya? Tahun berapakah kejadian tornado terbanyak?

```python
avgtornado = pd.read_sql_query("""SELECT COUNT(tornado) as total_tornado,
                                 AVG(wind_speed) AS AVERAGE_WIND_SPEED,
                                 AVG(TEMPERATURE) AS AVERAGE_TEMPERATURE
                                 FROM STATION_DATA
                                 WHERE TORNADO > 0 
                                 """,conn)
avgtornado
```

```python
tornadoyear = pd.read_sql_query("SELECT YEAR, SUM(TORNADO) AS TOTAL_TORNADO FROM STATION_DATA GROUP BY YEAR ORDER BY TOTAL_TORNADO DESC",conn)
tornadoyear
```
