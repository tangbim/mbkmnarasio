Case 4:<br>
a. Hitunglah persentase sales pada masing-masing status <br>
b. Hitunglah persentase cumulatif sales masing-masing status <br>
c. Buatlah pareto chart dan berikan intepretasinya

```python
# hitung total sales
total_sales = newdata['SALES'].sum()

# hitung persentase sales pada masing-masing status
sales_by_status = newdata.groupby('STATUS')['SALES'].sum()
sales_percentage = sales_by_status / total_sales * 100

print(sales_percentage)
```

```python
# hitung persentase cumulatif sales masing-masing status
cumulative_percentage = sales_percentage.cumsum()

print(cumulative_percentage)
```

```python
# buat pareto chart
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(sales_percentage.index, sales_percentage.values, color='tab:blue')
ax2.plot(sales_percentage.index, cumulative_percentage.values, color='tab:red', marker='o')
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.0%}".format(x/100)))
ax1.set_xlabel('Status')
ax1.set_ylabel('Sales Percentage')
ax2.set_ylabel('Cumulative Percentage')
ax1.tick_params(axis='x', rotation=45)

plt.show()
```
