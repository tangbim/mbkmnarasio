## Visualisasi Data

#### Bar Chart - Count Plot

Bar Chart digunakan untuk menunjukkan frekuensi atau nilai kuantitatif lainnya seperti rata-rata, jumlah, persentase dari data kategorik.

```python
len(df["region"].unique())
```

```python

# visualisasi banyaknya pencatatan tiap tahun
g = sns.catplot(x="region", kind="count", palette="flare", data=df)
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(20,8)
#plt.show()
```

Pencatatan data retail pada masing-masing region tidak jauh berbeda

#### Box Plot

Box plot digunakan untuk mengetahu penyebaran data. Ada banyak informasi yang kita dapatkan dari box plot yaitu nilai minimum, nilai maksimum, quartil pertama, median (quartil kedua), kuartil ketiga, dan outliers.

```python
# Visualisasi menggunakan seaborn
# visualisasi sebaran total volume setiap tahun berdasarkan type
g = sns.catplot(x="type", 
            y="Total Volume", hue="year", palette="flare", kind="box", data=df)
g.fig.set_size_inches(20,8)
plt.show()
```

```python
# Visualisasi menggunakan plotly
fig = px.box(df, x="type", y="Total Volume", color="year")
fig.show()
```

> Tipe konvensional memiliki lebih banyak outliers dibandingkan tipe organik. Terlihat nilai median tidak jauh berbeda setiap tahunnya pada masing-masing tipe, sedangkan nilai median pada conventional dan organic terlihat ada perbedaan.

#### Scatterplot

Scatterplot digunakan ketika ingin melihat relasi antara 2 variabel dan dari 2 variabel tersebut dapat diketahui variabel mana yang memiliki nilai yang lebih besar.

```python
# visualisasi hubungan variabel AveragePrice dan Total Volume
sns.scatterplot(x="AveragePrice",
                y="Total Volume",
                data = df)
plt.show()
```

```python
# memberikan warna yang berbeda untuk masing-masing type
plt.figure(figsize=(15,8))
ax = sns.scatterplot(x="AveragePrice",
                y="Total Volume",
                data = df,
                hue = 'type')

plt.show()
```

```python
# mengurutkan label type
plt.figure(figsize=(15,8))
ax = sns.scatterplot(x="AveragePrice",
                y="Total Volume",
                data = df,
                hue = 'type',
                hue_order=["organic","conventional"])
plt.show()
```

```python
# custom warna untuk setiap item di variabel type
hue_colors = {"organic":'green',
              "conventional":"red"}
sns.scatterplot(x="AveragePrice",
                y="Total Volume",
                data = df,
                hue = 'type',
                palette=hue_colors)
plt.show()
```

```python
# Visualisasi menggunakan plotly
fig = px.scatter(df, x="AveragePrice", y="Total Volume", color="type", color_discrete_sequence=['red', 'green'])
fig.update_traces(marker=dict(size=10,
                              line=dict(width=1,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```

> Total volume conventional lebih tinggi daripada total volume organik. Selain itu, rata-rata harga tipe organic tidak berhubungan linier terhadap total volume, semakin tinggi rata-rata harga tipe organic tidak menjadikan total volume semakin tinggi atau semakin rendah. Berbeda dengan tipe organic, rata-rata harga tipe convensional tampak mempengaruhi total volume.

### Pareto Chart

Pareto chart menunjukkan nilai kuantitatif dari data kategorik dengan menampilkannya secara terurut dari nilai tertinggi ke terendah. Selain itu juga menunjukkan persentase atau nilai kumulatif dari nilai kuantitatifnya.

```python
# menghitung rata-rata harga setiap region
data_group_cat=df.groupby('region')['AveragePrice'].mean()
```

```python
data_group_cat
```

```python
df_data_group_cat = data_group_cat.reset_index()
df_data_group_cat
```

```python
# menghitung persentase kumulatif dari AveragePrice
total_AveragePrice = df_data_group_cat.AveragePrice.sum()
df_data_group_cat['percentage_AveragePrice'] = df_data_group_cat.AveragePrice/total_AveragePrice
df_data_group_cat = df_data_group_cat.sort_values('percentage_AveragePrice', ascending = False)
df_data_group_cat['cumsum_pct'] = df_data_group_cat.percentage_AveragePrice.cumsum()
df_data_group_cat.index = df_data_group_cat.region
# df_data_group_cat = df_data_group_cat[df_data_group_cat.cumsum_pct <= 0.98] # untuk memfilter 98% averageprice yang ingin ditampilkan
df_data_group_cat = df_data_group_cat
df_data_group_cat.head()
```

```python
# visualisasi rata-rata harga alpukat pada setiap region
ax = df_data_group_cat.AveragePrice.plot(kind = 'bar',rot=90,color = 'r')
df_data_group_cat.cumsum_pct.plot(kind='line',
                                  linestyle='-',
                                  marker='o',figsize = (20,10),rot=90,secondary_y=True)
```

```python
# Pareto Chart menggunakan plotly
fig = go.Figure()

fig.add_trace(
    go.Bar(
        x= df_data_group_cat["region"],
        y= df_data_group_cat["AveragePrice"],
        name="Bar Chart"
    ))

fig.add_trace(
    go.Scatter(
        x= df_data_group_cat["region"],
        y= df_data_group_cat["cumsum_pct"]*df_data_group_cat["AveragePrice"].max(),
        mode="markers+lines",
        customdata = df_data_group_cat["cumsum_pct"],
        hovertemplate="cumsum_pct : %{customdata}",
        name="Cumulative"
    ))


fig.show()
```

Rata-rata harga alpukat tertinggi di region Hartford Springfield, San Francisco, dan Newyork.

### Tree Map

Tree map adalah jenis visualisasi yang membantu untuk melihat dan menampilkan informasi hirarkis dalam serangkaian persegi panjang yang berkerumun. Jika data Anda hirarki, Tree map adalah cara yang baik untuk menunjukkan semua nilai sekaligus serta dapat menjaga struktur data dalam bentuk visual

```python
fig = px.treemap(df, path=[px.Constant('Region'), 'region'], values='Total Volume', color='AveragePrice')
fig.show()
```

Terlihat dari Tree map di atas, Rata-rata harga alpukat tertinggi terdapat pada region yang memiliki warna kuning yang tinggi, yaitu di region Hartford Springfield, San Francisco, Newyork, dan Philadelphia.
