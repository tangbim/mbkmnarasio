Toko Bangunan “Maju Jaya” membuat laporan penjualan terhadap beberapa barang yang dijual selama tahun 2021. Berikut Tabel penjualan Toko Bangunan “Maju Jaya” selama tahun 2021.

(data berada pada soal)

1. Visualisasikan data tersebut ke dalam barplot! (Nilai Maks: 30)
2. Buatlah visualisasi  pareto chart dari data tersebut! (Nilai Maks: 46)

```python
# import package yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data preparation

df = pd.DataFrame({
      'Nama Barang': ['Keramik', 'Genteng', 'Kaca', 'Batu Bata', 'Besi'],
      'Kuantitas (unit)': [10000, 9500, 2000, 50000, 7500]
})

df.head()
```

1. Visualisasikan data tersebut ke dalam barplot! (Nilai Maks: 30)

```python
# membuat bar plot
plt.bar(df['Nama Barang'], df['Kuantitas (unit)'])
plt.title('Penjualan Barang Toko Maju Jaya Tahun 2021')
plt.xlabel('Nama Barang')
plt.ylabel('Kuantitas (unit)')
plt.show()
```

2. Buatlah visualisasi  pareto chart dari data tersebut! (Nilai Maks: 46)

```python
# membuat pareto chart
total = df['Kuantitas (unit)'].sum()
df = df.sort_values('Kuantitas (unit)', ascending=False)
df['Cumulative %'] = 100 * df['Kuantitas (unit)'].cumsum() / total

fig, ax1 = plt.subplots()

# membuat bar plot
ax1.bar(df['Nama Barang'], df['Kuantitas (unit)'], color='blue')
ax1.set_xlabel('Nama Barang')
ax1.set_ylabel('Kuantitas (unit)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Pareto Chart df Barang Toko Maju Jaya Tahun 2021')

# membuat line plot untuk cumulative percentage
ax2 = ax1.twinx()
ax2.plot(df['Nama Barang'], df['Cumulative %'], color='black', marker='o')
ax2.set_ylabel('Cumulative %', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# membuat grid line untuk cumulative percentage
ax2.grid(None)
for i in ax2.get_yticks():
      ax2.axhline(y=i, color='grey', alpha=0.4)

plt.show()
```
