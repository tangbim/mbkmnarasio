# Case 2

Apakah terdapat Missing Values ?

```python
jumlah_missing = pd.isnull(data).sum()
print(jumlah_missing)
print("\nYa, terdapat missing value pada kolom bike_id")
```

Apabila terdapat missing value, handling missing value tersebut.

```python
# mengecek rasio jumlah missing value
rasio = jumlah_missing.div(len(data))
print(rasio)
print("Kolom bike_id memiliki rasio <1% missing data sehingga akan dilakukan pengisian data")
```

```python
bikeiddesc = data['bike_id'].describe()
print(bikeiddesc)

# melakukan handling
data['bike_id'].fillna(int(data['bike_id'].mean()), inplace=True)
pd.isnull(data).sum()
```

# Case 4

Data pada kolom duration terdapat satuan waktu yang dapat menghalangi kita dalam menganalisis durasi pengendara sepeda. Oleh karena itu, dibutuhkan cleaning data satuan waktu.

```python
# melakukan trim string "minutes" agar tidak menghalangi kita dalam menganalisis
data['duration'] = data['duration'].str.replace(' minutes', '')
```

# Case 5

Karena tipe data duration seharusnya adalah integer, rubahlah tipe data duration menjadi integer.

```python
# mengubah duration dari string menjadi integer
data['duration'] = data['duration'].astype('int64')
data.info()
```

# Case 6

Hitunglah jumlah pengendara sepeda yang melewati masing-masing station A dan station B menggunakan groupby()

```python
bikersCount = data.groupby(['station_A_name', 'station_B_name'])['duration'].count()
bikersCount
```

Hitunglah jumlah pengendara sepeda berdasarkan user_type dan user_gender.

```python
byUserTypeGender = data.groupby(['user_type','user_gender'])['duration'].count()
byUserTypeGender
```

# Case 7

Berapakah rata-rata durasi dalam pemakaian sepeda?

```python
meanData = data['duration'].mean()
print(f"Rata-rata durasi dari pemakaian sepeda adalah {meanData:.1f} menit")
```

Berapakah rata-rata durasi pemakaian sepeda pada masing-masing rute?

```python
meanRouteDuration = data.groupby(['station_A_name', 'station_B_name'])['duration'].mean()
meanRouteDuration
```

Pemilik sepeda ingin memberikan promo kepada customer-nya yang bertujuan untuk meningkatkan ketertarikan pemuda maupun orang tua dalam bersepeda. Oleh karena itu pemilik sepeda ingin mengetahui rata-rata usia dan modus usia customer sehingga promo yang diberikan tepat.

```python
data['user_age'] = 2023 - data['user_birth_year']
meanAge = data['user_age'].mean()
modeAge = data['user_age'].mode()[0]
print(f"Rata-rata usia pengguna sepeda adalah {meanAge:.0f}")
print(f"modus usia pengguna sepeda adalah {modeAge}")
```
