```python
# 1
import pandas as pd

# declare data untuk dimasukkan datanya dari csv
data = pd.read_csv('./retail_raw_test.csv',index_col='order_date',parse_dates=True)
data.head(10)
```

```python
# 2
# menghapus kolom order_id dan product_id
data2 = data
data2 = data2.drop(columns=['order_id', 'product_id'])

# menampilkan data2 data2 pada kota Jakarta Pusat atau Yogyakarta
data2_jkt = data2[(data2['city'] == 'Jakarta Pusat') | (data2['city'] == 'Yogyakarta')]
print(data2_jkt)
```

```python
# 3
# membuat kolom total_price dan city/province
data2 = data
data2['total_price'] = data2['quantity'] * data2['item_price']
data2['city/province'] = data2['city'] + ' / ' + data2['province']

# menghapus kolom city dan province
data2 = data2.drop(columns=['city', 'province'])
print(data2.head())
```
