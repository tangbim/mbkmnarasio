# **Latihan (Probabilitas)**

Asosiasi  Olah Raga Nasional melakukan survei terhadap orang-orang berusia 7 tahun atau lebih tua tentang partisipasi dalam kegiatan olahraga (Abstrak Statistik Amerika Serikat: 2002).
Jumlah penduduk pada kelompok umur ini dilaporkan sebanyak 248,5 juta, dengan 120,9 juta laki-laki dan 127,6 juta perempuan. Jumlah peserta untuk lima aktivitas olahraga teratas adalah \\

|Activity| Male |Female|
|----|----|-----|
Bicycle riding |22.2| 21.0
Camping |25.6 |24.3
Exercise walking| 28.7 |57.7
Exercising with equipment |20.4 |24.4
Swimming |26.4| 34.4

a. Untuk wanita yang dipilih secara acak, perkirakan probabilitas partisipasi di masing-masing 5 kegiatan olahraga? \\
b. Untuk orang yang dipilih secara acak, berapa probabilitas orang tersebut berpartisipasi olahraga excercise walking?

```python
df = pd.DataFrame({
      'Activity' : ['Bicycle riding','Camping','Exercise walking', 
                  'Exercising with equipment','Swimming'],
      'Male' : [22.2,25.6,28.7,20.4,26.4],
      'Female' : [21.0,24.3,57.7,24.4,34.4] })

# Mengerjakan soal A
df['Probabilitas Wanita'] = df['Female'] / 127.6
df[['Activity','Probabilitas Wanita']]
```

```python
# Soal B
partisipan_exercise_walking = df.loc[df['Activity'] == 'Exercise walking', 'Male'].values[0] + df.loc[df['Activity'] == 'Exercise walking', 'Female'].values[0]
semesta = 248.5
exercise_walking_probabilitas = partisipan_exercise_walking / semesta
print(f"Jadi probabilitas orang yang melakukan exercise walking adalah {exercise_walking_probabilitas}")
```

# **Latihan (Statistika Deskriptif)**

Untuk lebih memahami materi statistika deskriptif, silakan untuk mencoba latihan ini

Untuk periode 2001â€“2008, Bristol-Myers Squibb
Company, Inc. melaporkan jumlah (dalam milyaran
dolar) untuk (1) penjualan bersih dan (2) iklan dan
promosi produk \\

| Tahun | Net sales | Iklan |
|----|---|----|
|2001 | 16.612 |1.201|
|2002 |16.208 |1.143|
|2003 |18.653 |1.416|
|2004 |19.380 |1.411|
|2005 |19.207 |1.476|
|2006 |16.208 |1.304|
|2007 |18.193 |1.415|
|2008 |20.597| 1.550|

(a) Tunjukkan berapa rata-rata net sales dan iklan selama 9 tahun? \\
(b) Berapa variasi yang didapatkan untuk net sales dan iklan selama 9 tahun? \\


```python
df2 = pd.DataFrame({
      'Tahun' : [2001,2002,2003,2004,2005,2006,2007,2008],
      'Net sales' : [16.612,16.208,18.653,19.380,19.207,16.208,18.193,20.597],
      'Iklan' : [1.201,1.143,1.416,1.411,1.476,1.304,1.415,1.550]
      })
print(df2)
# Tunjukkan berapa rata-rata net sales dan iklan selama 9 tahun?
mean_netsales = df2['Net sales'].mean()
mean_iklan = df2['Iklan'].mean()

print(f"Jadi rata-rata iklan adalah {mean_iklan}")
print(f"Jadi rata-rata net sales adalah {mean_netsales}")
```

```python
# B. Berapa variasi yang didapatkan untuk net sales dan iklan selama 9 tahun?
net_sales_var = df2['Net sales'].var()
iklan_var = df2['Iklan'].var()

print(f"Variansi net sales selama 9 tahun adalah {net_sales_var}")
print(f"Variansi iklan selama 9 tahun adalah {iklan_var}")
```

# Latihan (Statistika Deskriptif di bisnis)

Berdasarkan data transaksi dari coffee shop diatas. Carilah persebaran umur berdasarkan mean, dan varian dari customer tersebut. Serta dapatkan modus usia dari customers.

```python
# reading csv file
df3 = pd.read_csv('customer.csv')
# adding kolom usia
df3['Usia'] = 2023 - df3['birth_year']
df3
```

```python
# Carilah persebaran umur berdasarkan mean, dan varian dari customer tersebut. Serta dapatkan modus usia dari customers.

# Mean usia
meanage = df3['Usia'].mean()
print(f"Jadi usia rerata pada data tersebut adalah {meanage}")

# variasi usia
varage = df3['Usia'].var()
print(f"Jadi variasi usia pada data tersebut adalah {varage}")

# modus usia
modeage = df3['Usia'].mode()[0]
print(f"Jadi modus usia pada data tersebut adalah {modeage}")
```
