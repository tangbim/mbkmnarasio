## Pertanyaan 1
Mencari rata-rata dan standar deviasi dari luas area dan luas garasi yang ditawarkan.

```python
stdLotArea = data['LotArea'].std()
meanLotArea = data['LotArea'].mean()
print(f"Standar deviasi garasi berdasarkan data adalah {stdLotArea}")
print(f"Rata-rata luas garasi berdasarkan data adalah {meanLotArea}\n")

stdGarageArea = data['GarageArea'].std()
meanGarageArea = data['GarageArea'].mean()
print(f"Rata-rata luas garasi berdasarkan data adalah {meanGarageArea}")
print(f"Standar deviasi luas garasi berdasarkan data adalah {stdGarageArea}")
```

## Pertanyaan 2
Mencari rata-rata luas area dan garasi berdasarkan neighborhood dan street

```python
meanLotAreaGarageArea = data.groupby(['Neighborhood', 'Street']).mean()

res = meanLotAreaGarageArea[['LotArea', 'GarageArea']]

res
```

## Pertanyaan 3
Mencari kisaran luas area dan garasi untuk jalan dengan tipe pave

```python
# kisaran = range 
# rumus kisaran adalah data max - data min
def searchRange (data) :
      rangeKisaran = data.max() - data.min()
      return rangeKisaran

paveLotArea = data[data['Street']=='Pave']['LotArea']
rangeLotArea = searchRange(paveLotArea)

print(f"Jadi kisaran luas area dengan tipe pave adalah {rangeLotArea}")

paveGarageArea = data[data['Street']=='Pave']['GarageArea']
rangeGarageArea = searchRange(paveGarageArea)
print(f"Jadi kisaran luas area garasi dengan tipe pave adalah {rangeGarageArea}")
```
